# Etude sur Introduction aux réseaux de neurones artificiels - classification

La classification par un perceptron multi-couche sera illustrée dans trois Jupyter Notebooks sur l'exemple du Breast cancer wisconsin (diagnostic) dataset en utilisant exclusivement TensorFlow / Keras (pas de scikit-learn).

### 1 - Importation des librairies Python nécessaires à la résolution du problème

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense
from tensorflow.keras.optimizers import Adam
import pandas as pd
from joblib import load
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

import matplotlib.pyplot as plt
import tensorflow
from keras.callbacks import ModelCheckpoint
from tensorflow.keras.models import load_model
```

### 2 - affichage

```python
pd.set_option("max_columns", None)
pd.set_option("max_colwidth", None)
pd.set_option("max_row", 500)
```

### 3 - Chargement des données train

```python
X_train = load("X_train.joblib")
X_train.shape
```

```python
y_train = load("y_train.joblib")
y_train.shape
```

### 4 - Create a model

```python
# 1 couche caché est suffisant
# il y a 30 features => on prends le millieux pour units(neuron) pour commencer = 16
model = Sequential()

model.add(InputLayer(input_shape=(X_train.shape[1])))
model.add(Dense(units=16, activation='relu'))
#model.add(Dense(units=2, activation='softmax')) pour pleusieurs sorties activation='softmax'
model.add(Dense(units=1, activation='sigmoid')) # pour 1 sortie activation='sigmoid'
```

```python
#optimizer : Adam est plus utiliser en ce moment
model.compile(loss='binary_crossentropy',
              optimizer=Adam(learning_rate=0.1),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(
    x=X_train,
    y=y_train,
    batch_size=X_train.shape[0],
    epochs=100,
    validation_split=0.25,
    shuffle=True
)
```

```python
plt.title('Loss')
plt.plot(history.history['loss'], label='Train dataset')
plt.plot(history.history['val_loss'], label='Validation dataset')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.title('Accuracy')
plt.plot(history.history['accuracy'], label='Train dataset')
plt.plot(history.history['val_accuracy'], label='Validation dataset')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
y_train_pred = model.predict(X_train)
y_train_pred  = y_train_pred > 0.5

cm_train = confusion_matrix(y_train, y_train_pred)
cm_train
```

```python
print(classification_report(y_train, y_train_pred))
```

```python
accuracy_score_keras = tensorflow.keras.metrics.BinaryAccuracy()
accuracy_score_keras.update_state(y_train, y_train_pred)
print(accuracy_score_keras.result().numpy())
```

```python
#resultats = pd.DataFrame(columns=['units','learning_rate'])
columns2=['units','learning_rate','activation','Loss','val_loss','Accuracy','val_accuracy']
df1 = pd.DataFrame(columns=columns2)
learning_rate = [0.001,0.01,0.1]
i=[]
activation = ['relu','tanh']
epochs = 100
for i in range(2,15,1):
    for lr in learning_rate:
        for ac in activation:
            model = Sequential()
            model.add(InputLayer(input_shape=(X_train.shape[1])))
            model.add(Dense(units=i, activation=ac))
            model.add(Dense(units=1, activation='sigmoid'))

            model.compile(loss='binary_crossentropy',
                          optimizer=Adam(learning_rate=lr),
                          metrics=['accuracy'])

            modelCheckpoint = ModelCheckpoint(filepath = 'model_test',
                                      monitor='val_loss',
                                      verbose=1,
                                      save_best_only=True,
                                      save_weights_only=False,
                                      mode='min',
                                      #period=1
                                             )    

            history = model.fit(
                                x=X_train,
                                y=y_train,
                                batch_size=X_train.shape[0],
                                epochs=epochs,
                                validation_split=0.25,
                                shuffle=True,
                                callbacks=[modelCheckpoint]
                                )
            history
            
            hist = pd.DataFrame(history.history)
            loss = hist.loss[epochs-1]
            accuracy = hist.accuracy[epochs-1]
            val_loss = hist.val_loss[epochs-1]
            val_accuracy = hist.val_accuracy[epochs-1]
            
            df2 = pd.DataFrame(data=[[i,lr,ac,loss, val_loss,accuracy, val_accuracy]],columns=columns2 )
            df1 = df1.append(df2)
            print(df2)
print(df1)
#resultats = pd.DataFrame(i,columns=['units'])
#print('{}\n'.format(resultats))
```

### 6 - training the model with the best hyperparameter
Je vais mettre les meilleurs hyperparamettres

```python
model_best = Sequential()
model_best.add(InputLayer(input_shape=(X_train.shape[1])))
model_best.add(Dense(units=13, activation='relu'))
model_best.add(Dense(units=1, activation='sigmoid'))

model_best.compile(loss='binary_crossentropy',
                  optimizer=Adam(learning_rate=0.1),
                  metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping()
```

```python
history_best = model_best.fit(
    x=X_train,
    y=y_train,
    batch_size=X_train.shape[0],
    epochs=40,
    validation_split=0.25,
    #callbacks=[early_stopping],
    shuffle=True
)
```

```python
plt.title('Loss')
plt.plot(history_best.history['loss'], label='Train dataset')
plt.plot(history_best.history['val_loss'], label='Validation dataset')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.title('Accuracy')
plt.plot(history_best.history['accuracy'], label='Train dataset')
plt.plot(history_best.history['val_accuracy'], label='Validation dataset')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
