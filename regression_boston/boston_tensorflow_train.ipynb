```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense
from tensorflow.keras.optimizers import SGD
from joblib import load
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras.optimizers import Adam

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
X_train = load("X_train.joblib")
X_train.shape
```

```python
y_train = load("y_train.joblib")
y_train.shape
```

```python
model = Sequential()

model.add(InputLayer(input_shape=(X_train.shape[1])))
model.add(Dense(units=16, activation='relu'))
model.add(Dense(units=1, activation='linear'))
```

```python
model.compile(loss= 'mean_squared_error',
              optimizer=Adam(learning_rate=0.01),
              metrics=['mean_absolute_error']) # metrics = ['mse'] aussi bon
```

```python
model.summary()
```

```python
history = model.fit(
    x=X_train,
    y=y_train,
    batch_size=X_train.shape[0],
    epochs=200,
    validation_split=0.25,
    shuffle=True
)
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
hist.describe()
```

```python
plt.title('Loss')
plt.plot(history.history['loss'], label='Train dataset')
plt.plot(history.history['val_loss'], label='Validation dataset')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.title('mean_absolute_error')
plt.plot(history.history['mean_absolute_error'], label='Train dataset')
plt.plot(history.history['val_mean_absolute_error'], label='Validation dataset')
plt.xlabel('Epoch')
plt.ylabel('mean_absolute_error')
#plt.ylim([0,5])
plt.legend()
plt.show()
```

```python
# print score
y_train_pred = model.predict(X_train).flatten()
print('R2 score: {0:0.2%}'.format(r2_score(y_train, y_train_pred)))
print('RMSE: {}'.format(mean_squared_error(y_train, y_train_pred, squared=False)))
```

```python
model.save('model.h5')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping()
```
